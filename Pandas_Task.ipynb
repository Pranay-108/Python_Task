# Pandas
The Pandas library in Python is a powerful and versatile tool for data manipulation and analysis. It provides data structures like Series and DataFrame, which make it easy to clean, analyze, and visualize data.

```python
import pandas as pd
```

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 23, 34, 29],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df = pd.DataFrame(data)
print(df)
```

## Sales Data

```python
Sales = pd.read_csv("D:/Data Analysis/Mongo DB/sales.csv",index_col=0,header=0)
```

```python
Sales50 = Sales.head(50)
```

```python
Sales50.to_dict(orient='records')
```

```python
SalesDF = pd.DataFrame(Sales50)
SalesDF
```

```python
SalesDF.describe()
```

```python
SalesDF.to_csv('SalesDF.csv', index=False)
```
